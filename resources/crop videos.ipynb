```python
import imageio
from glob import glob
import os.path as osp
import matplotlib.pyplot as plt
import numpy as np
import subprocess

from tqdm.auto import tqdm
```

```python
def bbox(img):
    rows = np.any(img, axis=1)
    cols = np.any(img, axis=0)
    rmin, rmax = np.where(rows)[0][[0, -1]]
    cmin, cmax = np.where(cols)[0][[0, -1]]

    return cmin, rmin, cmax, rmax
```

```python
def crop_video(input_path, output_path, bbox):
    x1, y1, x2, y2 = bbox
    cmd = [
        "ffmpeg",
        "-i",
        input_path,
        "-filter:v",
        f"crop={x2-x1}:{y2-y1}:{x1}:{y1}",
        "-y",
        output_path,
    ]
    subprocess.call(cmd)
```

```python
video_dir = "self_captures"
output_dir = "self_captures_cropped"
```

```python
video_paths = sorted(glob(osp.join(video_dir, "*.mp4")))
```

```python
deltas = []
for path in tqdm(video_paths):
    min_y = np.inf
    max_y = 0
    for frame in imageio.imiter(path):
        mask = np.all(frame < 240, axis=2)
        x1, y1, x2, y2 = bbox(mask)
        min_y = min(y1, min_y)
        max_y = max(y2, max_y)

    delta = max_y - min_y
    deltas.append(delta)
```

```python
image_size = 1024
d = 156
d = 350
```

```python
deltas = []
for path in tqdm(video_paths):
    min_y = np.inf
    max_y = 0
    for frame in imageio.imiter(path):
        mask = np.all(frame < 240, axis=2)
        x1, y1, x2, y2 = bbox(mask)
        min_y = min(y1, min_y)
        max_y = max(y2, max_y)

    m = (min_y + max_y) / 2
    
    start = max(m - d, 0)
    end = start + 2 * d
    
    new_bbox = [0, start, image_size, end]
    
    new_path = osp.join(output_dir, osp.basename(path))
    crop_video(path, new_path, new_bbox)
```
